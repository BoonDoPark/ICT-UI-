### 1.2. 다중 그래프 (multiple graphs)

1개의 canvas 안에 다중 그래프 그리기

```python
data = np.arange(1, 51)
plt.plot(data)

data2 = np.arange(51, 101)
plt.plot(data2)

plt.plot(data2+50)

plt.show()
```

2개의 figure로 나누어서 다중 그래프 그리기
* figure()는 새로운 그래프 canvas를 생성합니다.

```python
data = np.arange(100, 201)
plt.plot(data)

data2 = np.arange(200, 301)
plt.figure()
plt.plot(data2)


plt.show()
```

### 1.3. 여러개의 plot을 그리는 방법 (subplot)

subplot(row, column, index)

```python
data = np.arange(100, 201)
plt.subplot(2, 1, 1)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(2, 1, 2)
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100, 201)
plt.subplot(1, 3, 1)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(1, 3, 2)
plt.plot(data2)

data3 = np.arange(300, 401)
plt.subplot(1, 3, 3)
plt.plot(data3)

# 조금 더 타이트하고 이쁘게 출력되는 옵션
plt.tight_layout()
plt.show()
```

위의 코드와 동일하나 , (콤마)를 제거한 상태

```python
data = np.arange(100, 201)
plt.subplot(211)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(212)
plt.plot(data2)


plt.show()
```

```python
data = np.arange(100, 201)
plt.subplot(2, 3, 1)
plt.plot(data)
plt.subplot(2, 3, 2)
plt.plot(data)
data2 = np.arange(200, 301)
plt.subplot(2, 3, 3)
plt.plot(data2)
plt.subplot(2, 3, 4)
plt.plot(data2)
data3 = np.arange(300, 401)
plt.subplot(2, 3, 5)
plt.plot(data3)
plt.subplot(2, 3, 6)
plt.plot(data3)

plt.tight_layout()
plt.show()
```

### 1.4. 여러개의 plot을 그리는 방법 (subplots) - s가 더 붙습니다.

plt.subplots(행의 갯수, 열의 갯수)

```python
data = np.arange(1, 51)

# 밑그림
fig, axes = plt.subplots(2, 3)

axes[0, 0].plot(data)
axes[0, 1].plot(data*data)
axes[0, 2].plot(data**3)
axes[1, 0].plot(data%10)
axes[1, 1].plot(-data)
axes[1, 2].plot(data//20)

plt.tight_layout()
plt.show()
```

### 2.1. 타이틀

```python
plt.plot([1, 2, 3], [3, 6, 9])
plt.plot([1, 2, 3], [2, 4, 9])

# 한글 폰트 설정
# 맑은 고딕 : Malgun Gothic, 나눔고딕 : Nanum Gothic
font_test = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_test).get_name()
print(font_name)
rc('font', family=font_name)

# 타이틀 지정
plt.title('이것은 타이틀입니다.')

plt.show()
```

```python
plt.plot([1, 2, 3], [3, 6, 9])
plt.plot([1, 2, 3], [2, 4, 9])

# 타이틀 : font size
plt.title('타이틀의 fontsize를 키웁니다.', fontsize=20)


plt.show()
```

### 2.2. X, Y축 Label 설정

```python
plt.plot([1, 2, 3], [3, 6, 9])
plt.plot([1, 2, 3], [2, 4, 9])

# 타이틀 : font size
plt.title('타이틀의 fontsize를 키웁니다.', fontsize=20)

# x, y축 label 설정
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)

plt.show()
```

### 2.3. X, Y 축 Tick 설정 (rotation)

Tick은 X, Y축에 위치한 눈금을 말합니다.

```python
plt.plot(np.arange(1, 11), np.arange(1, 11)*2)
plt.plot(np.arange(1, 11), np.arange(1, 11)**2)
plt.plot(np.arange(1, 11), np.log(np.arange(1, 11)))


# 타이틀 : font size
plt.title('타이틀의 fontsize를 키웁니다.', fontsize=20)

# x, y축 label 설정
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)

# x tick, y tick 척도결정
plt.xticks(rotation=90)
plt.yticks(rotation=30)


plt.show()
```

### 2.4. 범례(Legend) 설정 

```python
plt.plot(np.arange(1, 11), np.arange(1, 11)*2)
plt.plot(np.arange(1, 11), np.arange(1, 11)**2)
plt.plot(np.arange(1, 11), np.log(np.arange(1, 11)))


# 타이틀 : font size
plt.title('타이틀의 fontsize를 키웁니다.', fontsize=20)

# x, y축 label 설정
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)

# x tick, y tick 척도결정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

# legend 설정
plt.legend(['10*2', '10**2', 'log'], fontsize=15, loc=1)

plt.show()
```

### 2.5. X와 Y의 한계점(Limit) 설정

```python
plt.plot(np.arange(1, 11), np.arange(1, 11)*2)
plt.plot(np.arange(1, 11), np.arange(1, 11)**2)
plt.plot(np.arange(1, 11), np.log(np.arange(1, 11)))


# 타이틀 : font size
plt.title('타이틀의 fontsize를 키웁니다.', fontsize=20)

# x, y축 label 설정
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)

# x tick, y tick 척도결정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

# legend 설정
plt.legend(['10*2', '10**2', 'log'], fontsize=15, loc=1)

# xlim, ylim 설정
plt.xlim(0, 5)
plt.ylim(0, 5, 10)

plt.show()
```

### 2.6. 스타일 세부 설정 - 마커, 라인, 컬러

스타일 세부 설정은 마커, 선의 종류 설정, 그리고 컬러가 있으며, 문자열로 세부설정을 하게 됩니다.

[세부 도큐먼트 확인하기](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

**marker의 종류**
* '.'	point marker
* ','	pixel marker
* 'o'	circle marker
* 'v'	triangle_down marker
* '^'	triangle_up marker
* '<'	triangle_left marker
* '>'	triangle_right marker
* '1'	tri_down marker
* '2'	tri_up marker
* '3'	tri_left marker
* '4'	tri_right marker
* 's '	square marker
* 'p'	pentagon marker
* '*'	star marker
* 'h'	hexagon1 marker
* 'H'	hexagon2 marker
* '+'	plus marker
* 'x'	x marker
* 'D'	diamond marker
* 'd'	thin_diamond marker
* '|'	vline marker
* '_'	hline marker

```python
plt.plot(np.arange(1,11), np.arange(1,11)*2, marker='o', markersize=10)
plt.plot(np.arange(1,11), np.arange(1,11)*2-10, marker='v', markersize=10)
plt.plot(np.arange(1,11), np.arange(1,11)*2-20, marker='d', markersize=10)
plt.plot(np.arange(1,11), np.arange(1,11)*2-30, marker='*', markersize=10)

# 타이틀 & 폰트 사이즈
plt.title('타이틀', fontsize= 30)

# x축, y축 labal 설정
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)

# xtick, ytick 회전
plt.xticks(rotation=90)
plt.yticks(rotation=90)

# legend 활용
plt.legend(['x2', 'x2-10', 'x2-20', 'x2-30'])

plt.show()
```

**line의 종류**
* '-' solid line style
* '--' dashed line style
* '-.' dash-dot line style
* ':' dotted line style

```python
plt.plot(np.arange(1,11), np.arange(1,11)*2, marker='o', linestyle='-')
plt.plot(np.arange(1,11), np.arange(1,11)*2-10, marker='v', linestyle='--')
plt.plot(np.arange(1,11), np.arange(1,11)*2-20, marker='d', linestyle='-.')
plt.plot(np.arange(1,11), np.arange(1,11)*2-30, marker='*', linestyle=':')
plt.plot(np.arange(1,11), np.arange(1,11)*2-40, marker='h', linestyle='')

# 타이틀 & 폰트 사이즈
plt.title('타이틀', fontsize= 30)

# x축, y축 labal 설정
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)

# xtick, ytick 회전
plt.xticks(rotation=90)
plt.yticks(rotation=90)

# legend 활용
plt.legend(['x2', 'x2-10', 'x2-20', 'x2-30'])

plt.show()
```

**color의 종류**
* 'b'	blue
* 'g'	green
* 'r'	red
* 'c'	cyan
* 'm'	magenta
* 'y'	yellow
* 'k'	black

```python
plt.plot(np.arange(1,11), np.arange(1,11)*2, marker='o', linestyle='-', color='cyan')
plt.plot(np.arange(1,11), np.arange(1,11)*2-10, marker='v',linestyle='--', color='blue')
plt.plot(np.arange(1,11), np.arange(1,11)*2-20, marker='d',linestyle='-.', color='red')
plt.plot(np.arange(1,11), np.arange(1,11)*2-30, marker='*',linestyle=':', color='black')
plt.plot(np.arange(1,11), np.arange(1,11)*2-40, marker='H', linestyle='', color='green')
# 타이틀 & 폰트 사이즈
plt.title('타이틀', fontsize= 30)

# x축, y축 labal 설정
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)

# xtick, ytick 회전
plt.xticks(rotation=90)
plt.yticks(rotation=90)

# legend 활용
plt.legend(['x2', 'x2-10', 'x2-20', 'x2-30'])

plt.show()
```
